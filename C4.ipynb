Que pasa? me di cuenta que con .split() cuenta menos trabajos con la palabra chief que cuando lo buscas sin el split.
Hay trabajos que tienen la palabra chief seguido de una coma sin espacio, entonces al poner split te toma la palabra entera "chief," y no la cuenta como "chief"

con el siguiente codigo se puede ver la diferencia

```python
count=0
for row in sal['JobTitle']:
    if 'chief' in row.lower().split():  
        count=count+1
        print (row.lower().split(),count)
```

```python
count=0
for row in sal['JobTitle']:
    if 'chief' in row.lower():  #sin .split() cuenta una suma mayor
        count=count+1
        print (row.lower(),count)
```
